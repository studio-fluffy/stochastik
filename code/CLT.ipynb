```r
Outcome <- sample(1:6,10000, replace= TRUE)
hist(Outcome, col ="light blue")
abline(v=3.5, col = "red",lty=1)
```

```r
x10 <- c()
sample_size = 5000
k =10000
for ( i in 1:k) {
 x10[i] = mean(sample(1:6,sample_size, replace = TRUE))
}

hist(x10, col ="pink", main="Sample size", xlab ="Outcome of die roll")
abline(v = mean(x10), col = "Red")
abline(v = 3.5, col = "blue")
```
